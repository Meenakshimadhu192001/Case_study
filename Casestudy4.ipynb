```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import time

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import tensorflow as tf
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Cell C: handle missing values (THIS MUST COME BEFORE ENCODING)
df['income'] = df['income'].str.replace('.', '', regex=False)
df.replace('?', np.nan, inplace=True)
df.dropna(inplace=True)

print(df.isnull().sum().sum())
print(df['income'].unique())
```

```python
df.drop_duplicates(inplace=True)
```

###Encoding

```python
cat_cols = df.select_dtypes(include='object').columns

print(cat_cols)
```

```python

cat_cols = df.select_dtypes(include='object').columns
le = LabelEncoder()
for col in cat_cols:
    df[col] = le.fit_transform(df[col])
```

###Split Features and Target

```python
#splitting to target and features
y=df['income']            #target
x=df.drop(columns=['income'],axis=1) #features
#splitting into test train data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=33)
```

###scaling

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

```python
# TensorFlow model
tf_model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(x_train.shape[1],)),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
tf_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
start_time = time.perf_counter()
history = tf_model.fit(x_train, y_train, epochs=50, batch_size=32,
                        validation_split=0.2, verbose=1)
end_time = time.perf_counter()
print(f"Training Time: {end_time - start_time:.4f} seconds")
```

```python
loss, accuracy = tf_model.evaluate(x_test, y_test)
print("accuracy:", accuracy)
```

##PyTorch

```python
# Cell 22 correction
x_train_tensor = torch.FloatTensor(X_train)
x_test_tensor = torch.FloatTensor(X_test)
y_train_tensor = torch.FloatTensor(y_train.values.reshape(-1,1))
y_test_tensor = torch.FloatTensor(y_test.values.reshape(-1,1))
```

```python
import torch
import torch.nn as nn

class ANN(nn.Module):
    def __init__(self, input_dim):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.network(x)
```

```python
model = ANN(input_dim=x_train_tensor.shape[1])
```

```python
criterion = nn.BCELoss()
```

```python
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
start_time = time.perf_counter()
epochs = 50
for epoch in range(epochs):
  outputs = model(x_train_tensor)
  loss = criterion(outputs, y_train_tensor)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
  if (epoch+1) % 10 == 0:
    print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
end_time = time.perf_counter()
print(f"Training time: {end_time-start_time:.4f}seconds")
```
